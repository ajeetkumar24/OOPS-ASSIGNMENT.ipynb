**Method Overloading:**

Two or more methods have the same name but different numbers of parameters or different types of parameters, or both. These methods are called overloaded methods and this is called method overloading.


The problem with method overloading in Python is that we may overload the methods but can only use the latest defined method.








```python
# First product method.
# Takes two argument and print their
# product


def product(a, b):
    p = a * b
    print(p)

# Second product method
# Takes three argument and print their
# product


def product(a, b, c):
    p = a * b*c
    print(p)

# Uncommenting the below line shows an error
# product(4, 5)


# This line will call the second product method
product(4, 5, 5)
```

In the above code, we have defined two product methods we can only use the second product method, as python does not support method overloading. We may define many methods of the same name and different arguments, but we can only use the latest defined method. Calling the other method will produce an error. Like here calling product(4,5) will produce an error as the latest defined product method takes three arguments.

Thus, to overcome the above problem we can use different ways to achieve the method overloading.

**Single Inheritance:**

Single inheritance enables a derived class to inherit properties from a single parent class, thus enabling code reusability and the addition of new features to existing code.







image.png

```python
# Python program to demonstrate
# single inheritance

# Base class
class Parent:
	def func1(self):
		print("This function is in parent class.")

# Derived class


class Child(Parent):
	def func2(self):
		print("This function is in child class.")


# Driver's code
object = Child()
object.func1()
object.func2()
```

**2.Multiple Inheritance:**

When a class can be derived from more than one base class this type of inheritance is called multiple inheritances. In multiple inheritances, all the features of the base classes are inherited into the derived class.


image.png

```python
# Python program to demonstrate
# multiple inheritance

# Base class1
class Mother:
	mothername = ""

	def mother(self):
		print(self.mothername)

# Base class2


class Father:
	fathername = ""

	def father(self):
		print(self.fathername)

# Derived class


class Son(Mother, Father):
	def parents(self):
		print("Father :", self.fathername)
		print("Mother :", self.mothername)


# Driver's code
s1 = Son()
s1.fathername = "Ramashish Prasad"
s1.mothername = "Shiv kumari devi"
s1.parents()
```

**3.Multilevel Inheritance :**

In multilevel inheritance, features of the base class and the derived class are further inherited into the new derived class. This is similar to a relationship representing a child and a grandfather.


image.png


```python
class ParentClass1:
    def method1(self):
        print("method 1 of parent class 1")
class ParentClass2:
    def method2(self):
        print("method2 of parent class 2")
```

```python
class ChildClass(ParentClass1, ParentClass2):
    def method(self):
        print("method of child class")
```

```python
c1 = ChildClass()
c1.method()
```

When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance. In this program, we have a parent (base) class and two child (derived) classes.

```python
#7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?
```

**Method Resolution Order :**

Method Resolution Order(MRO) it denotes the way a programming language resolves a method or attribute. Python supports classes inheriting from other classes. The class being inherited is called the Parent or Superclass, while the class that inherits is called the Child or Subclass. In python, method resolution order defines the order in which the base classes are searched when executing a method. First, the method or attribute is searched within a class and then it follows the order we specified while inheriting. This order is also called Linearization of a class and set of rules are called MRO(Method Resolution Order). While inheriting from another class, the interpreter needs a way to resolve the methods that are being called via an instance. Thus we need the method resolution order. For Example








In Python, the Method Resolution Order (MRO) defines the order in which Python looks for a method in a hierarchy of classes, especially with multiple inheritance. You can access the MRO of a class using the __mro__ attribute or the mro() method

```python
class A:
    def method(self):
        print("A method")

class B:
    def method(self):
        print("B method")

class C(A, B):
    pass

c = C()
c.method()

print(C.__mro__)
```

**Explanation:**

* In this example, class C inherits from both A and B.
When c.method() is called, Python uses the MRO to determine which method() to execute.

* The MRO for C is (C, A, B, object), which means Python will first look for method() in C, then in A, then in B, and finally in the base class object.

* Since C doesn't define its own method(), it finds method() in A and executes it.

Create an abstract base class called Shape with an abstract method area(). Then, create concrete subclasses of Shape, such as Circle and Rectangle, and implement the area() method in these subclasses.


**Abstract classes in Python**

Abstract classes define common behavior, which ensures that all subclasses have the required methods. This allows objects of different descendants of an abstract base class to be used interchangeably.


```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        """
        Abstract method to calculate the area of a shape.
        Must be implemented by subclasses.
        """
        pass
```

**Explanation:**

* from abc import ABC, abstractmethod:

This line imports the necessary components to create an abstract base class (ABC) in Python.

* class Shape(ABC):

Defines a class named "Shape" that inherits from the ABC class, making it an abstract base class.

* @abstractmethod:
This decorator marks the area method as an abstract method, meaning that any subclass of Shape must implement this method to be considered concrete.

* def area(self):
This is the definition of the abstract area method. It does not have any implementation as it is meant to be overridden by subclasses.


```python
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

# Example usage
square = Square(5)
print(square.area())

circle = Circle(3)
print(circle.area())
```

To implement encapsulation in a BankAccount class, you can use private attributes for balance and account_number, and include methods for deposit, withdrawal, and balance inquiry:

**Attributes**

Use an underscore to mark the attributes as private, indicating that they should not be accessed directly from outside the class.

**Methods**

Include public methods to perform operations on the bank account, such as depositing or withdrawing money and checking the account balance.

**Encapsulation**

Encapsulation is the process of integrating data and the code that acts upon it as a single unit. In this example, the internal details of managing the balance and account number are private from users, and only the necessary methods are exposed.

Here's an example of a BankAccount class in Python that implements encapsulation:


```python
class BankAccount:
  def __init__(self, account_number, balance):
    self._account_number = account_number
    self._balance = balance
  def deposit(self, amount):
    self._balance += amount
    return self._balance
  def withdraw(self, amount):
    if amount <= self._balance:
      self._balance -= amount
      return self._balance
    else:
      return "Insufficient balance"
  def get_balance(self):
    return self._balance
```

```python
#decorator approach for functions>>use case 1
def my_decorator(func): #decorator function that takes another func as arguement
    def wrapper(): #wrapper is a nested function that adds the functionality before and after calling original func func
        print("The line before computation.")
        func() #say_hello which is the func here will be executed here
        print("The line after computation")
    return wrapper
```

```python
@my_decorator #syntax to decorate the say hello function
def say_hello():
    print("Hello")
```

```python
#another use case of function decorator
#run time of a code

import time
def timer_decorator(func):
    def timer():
        start = time.time()
        func()
        end = time.time()
        print("The time for executing the code", end-start)
    return timer
```

```python
@timer_decorator
def func_test():
    print(1100000*1000)
```

```python
#15. Implement a static method in a class that checks if a given year is a leap year.
```

```python
public class DateUtil {

    public static boolean isLeapYear(int year) {
        return (year % 4 == 0 && year % 100 != 0) || (year % 400 == 0);
    }
}
```

**Explanation:**

Class Definition:

This code creates a class called DateUtil which will hold our static method to check for leap years.


**isLeapYear method:**

* Static: The public static keyword indicates that this method belongs to the class itself and can be called without creating an instance of the DateUtil class.

* Parameter: The method takes an integer year as input, representing the year to be checked.

* Logic:
  * The core logic is within the return statement:

  *    (year % 4 == 0 && year % 100 != 0): This checks if the year is divisible by 4 but not by 100, which is the primary condition for a leap year.

  * || (year % 400 == 0): This is an additional check for century years (divisible by 100) that are still considered leap years if they are also divisible by 400.

* Return Value: The method returns a boolean value: true if the given year is a leap year, false otherwise

```python
int yearToCheck = 2024;
boolean isLeap = DateUtil.isLeapYear(yearToCheck);
System.out.println(isLeap); // This will print "true" because 2024 is a leap year
```
